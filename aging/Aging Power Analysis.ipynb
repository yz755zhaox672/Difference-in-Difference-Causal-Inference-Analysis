## Aging power analysis 

## Notebook description

In this notebook we estimate the minimum number of patients required in each group for Aging to have 80% power.

```python
from numpy import array
from matplotlib import pyplot
from statsmodels.stats.power import TTestIndPower
import pandas as pd
import numpy as np
import os, sys
scripts_master = os.path.dirname('../../advanced_stats_code_library/src/')
sys.path.insert(0, scripts_master) 
from power_analysis.power_analysis import PowerAnalysis
```

```python
power_analysis=PowerAnalysis()
```

## Power analysis (per patient)

```python
effect_sizes = array([0.6, 0.7, 0.8])
sample_sizes = array(range(5, 146))
analysis = TTestIndPower()
analysis.plot_power(dep_var='nobs', nobs=sample_sizes, effect_size=effect_sizes)
pyplot.show()
```

### Based on domain knowledge from CASE study and WAM study.

CASE study effect size: 0.62  
WAM study effect size: 0.60

Effect size between 0.60 and 0.70 can be considered as a descent domain knowledge starting point to estimate the effect size for Aging study.

```python
power_analysis.power_ttestind(effect_size=0.62, power_req=0.80, nobs_range=(2,130), sig_level=0.05)
```

```python
power_analysis.power_ttestind(effect_size=0.70, power_req=0.80, nobs_range=(2,130), sig_level=0.05)
```

## Binomial test power analysis (based on WAM)

We get 0.76 and 0.42 as the proportions from WAM responder analysis using threshold method (from group of means). Using these proportions we get 0.70 as the proportion effect size. For these proportions, the minimum number of samples required per group is approximately ~28 based on the binomial power analysis.

```python
power_analysis.get_proportion_effectsize(0.76,0.42)
```

```python
power_analysis.get_proportion_effectsize(0.57,0.23)
```

```python
power_analysis.power_binomial(0.76,0.42)
```

```python
power_analysis.power_binomial(0.57,0.23)
```
