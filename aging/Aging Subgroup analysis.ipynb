# Subgroup Analysis

Subgroup analysis is performed within the ITT population during the baseline period to identify variables for potential stratification or adjustment.  The mean and standard deviation, as well as median and interquartile range (IQR), of the daily duration of each behavior during the baseline period will be calculated for population subgroups (without regard to study group) and statistical tests to identify behaviors that differ significantly across groups are performed.

```python
def age_category_converter(age):
    if (age < 7):
        return "Adult"
    elif (age>=7 and age < 11):
        return "Senior younger than 11"
    elif (age>= 11):
        return "Senior older than 11"
    else:
        return np.nan

def body_weight_category(x):
    if x is np.nan:
        return np.nan
    elif x>0 and x<=20:
        return 'up to 20 lbs'
    elif x>20 and x<=50:
        return '>20 lbs & <=50 lbs'
    elif x>50:
        return '>50 lbs'
    
def diary_category_converter(df):
    # Calculate the mean and median of 'Owner Daily Diary'
    mean_diary = df['Owner Daily Diary'].mean()
    median_diary = df['Owner Daily Diary'].median()

    # Create a new column 'Owner Daily Diary,mean.categorical'
    df['Owner Daily Diary.mean.categorical'] = df['Owner Daily Diary'].apply(
        lambda x: 'Above Mean' if x > mean_diary else 'Below Mean'
    )

    df['Owner Daily Diary.median.categorical'] = df['Owner Daily Diary'].apply(
        lambda x: 'Above Median' if x > median_diary else 'Below Median'
    )

def map_weekend_weekday(value):
    if value == 0:
        return 'Weekday'
    elif value == 1:
        return 'Weekend'
```

```python
def calculate_statistics(df, column_name, subgroup_column):
    grouped = df.groupby(subgroup_column)
    statistics = grouped[column_name].agg(['mean', 'std', 'median', lambda x: x.quantile(0.75) - x.quantile(0.25)])
    
    statistics.columns = ['Mean', 'SD', 'Median', 'IQR']

    # Round the statistics to 2 decimal places
    statistics_rounded = statistics.round(2)

    return statistics_rounded
```

```python
def statistical_tests(df, column_name, subgroup_column):

    grouped_data = df.groupby(subgroup_column)[column_name]
    
    # Kruskal-Wallis test for multiple subgroups
    kruskal_stat, p_value_kruskal = stats.kruskal(*[grouped_data.get_group(group) for group in df[subgroup_column].unique()])
    
    # Pairwise Mann-Whitney U test for two subgroups
    subgroup_list = df[subgroup_column].unique()
    pairwise_tests = []
    for i in range(len(subgroup_list)):
        for j in range(i+1, len(subgroup_list)):
            group1_data = grouped_data.get_group(subgroup_list[i])
            group2_data = grouped_data.get_group(subgroup_list[j])
            u_stat, p_value_mannwhitney = stats.mannwhitneyu(group1_data, group2_data, alternative='two-sided')
            pairwise_tests.append((subgroup_list[i], subgroup_list[j], u_stat, p_value_mannwhitney))
    
    print(f"Kruskal-Wallis Test Results for '{column_name}' across '{subgroup_column}':")
    print(f"Kruskal-Wallis Statistic: {kruskal_stat:.4f}")
    print(f"P-Value: {p_value_kruskal:.4f}\n")

    #print("Pairwise Mann-Whitney U Test Results:")
    #for group1, group2, u_stat, p_value in pairwise_tests:
       # print(f"{group1} vs. {group2}:")
       # print(f"Mann-Whitney U Statistic: {u_stat:.4f}")
       # print(f"P-Value: {p_value:.4f}")
       # print()
```

```python
def statistical_tests(df, column_name, subgroup_column):
    grouped_data = df.groupby(subgroup_column)[column_name]
    
    # Kruskal-Wallis test for multiple subgroups
    kruskal_stat, p_value_kruskal = stats.kruskal(*[grouped_data.get_group(group) for group in df[subgroup_column].unique()])

    # Formatting Kruskal-Wallis test results
    kruskal_data = [[kruskal_stat, p_value_kruskal]]
    kruskal_headers = ["Statistic", 'P-value']
    kruskal_stubs = ["Kruskal-Wallis Test"]
    tbl_kruskal = SimpleTable(kruskal_data, kruskal_headers, kruskal_stubs,
                              title=f"Kruskal-Wallis Test Results for '{column_name}' across '{subgroup_column}'")
    print(tbl_kruskal)
    
    # Conclusion based on p-value
    if p_value_kruskal < 0.05:
        conclusion = 'We reject the null hypothesis. At least one group median is different from others.'
    else:
        conclusion = 'We could not reject the null hypothesis. The medians of all groups are equal.'

    print(conclusion)
```
